```python

def remove_symbols(text, sentiments):
    clean_text = []
    clean_sentiments = []
    indices = []
    for i, (row, sentiment) in enumerate(zip(text, sentiments)):
        clean_row = re.sub(r'[^\w\s]', "", row)
        if clean_row.strip():  
            clean_text.append(clean_row)
            clean_sentiments.append(sentiment)
            indices.append(i)
    return clean_text, clean_sentiments, indices


```

```python
aft_re, clean_sentiments, indices = remove_symbols(data2__text, data2_sentiments)
```

```python
from nltk.corpus import stopwords as sp
```

```python
stop_words = set(sp.words('english'))
```

```python
def stopword(text):
 cleaned=[]
 for sentence in text :
   words =wtk(sentence)
   filtered = [word for word in words if word not in stop_words]
   cleaned.append(" ".join (filtered))
 return cleaned  
```

```python
aft_stopwords=stopword(aft_re)
```

```python
ps=PorterStemmer()
lem=WordNetLemmatizer()
```

```python
aft_stopwords
```

```python
def steming(text):
     filtered=[]
     for sentence in text:
          words=(wtk(sentence))
          stemmed=[ps.stem(word) for word in words ]
          filtered.append(' '.join(stemmed))
     return filtered
```

```python
aft_stemming=steming(aft_stopwords)
```

```python
aft_stemming
```

```python
def lemmatizer (text):
    filtered=[]
    for sentence in text:
          words=(wtk(sentence))
          lemmatized=[lem.lemmatize(word) for word in words ]
          filtered.append(' '.join(lemmatized))
    return filtered    
```

```python
aft_lemmatize= lemmatizer(aft_stemming)
```
